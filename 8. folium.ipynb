```python
import folium
```

```python
# folium.Map(location=[위도, 경도], zoom_start=배율, ...)
map_folium = folium.Map(location=[37.50006,127.0355806], zoom_start=17)
map_folium
```

```python
# Marker() : location 속성으로 지정한 위치에 마커를 생성
# popup() : 마커를 클릭했을 때 표시할 문자열을 생성
# add_to() : 작성된 마커를 지도에 추가

map_folium = folium.Map(location=[37.50006,127.0355806], zoom_start=17)
popup = folium.Popup("코리아IT아카데미", max_width=200)
folium.Marker(location=[37.50006,127.0355806], popup=popup).add_to(map_folium)
map_folium
```

```python
# 지도: 37.56652479270633, 126.9779129116262
# 서울특별시청: 37.56652479270633, 126.9779129116262
# 덕수궁: 37.56585320879458, 126.97514572197248
# 서울지방경찰청: 37.57504045678939, 126.97195576938117
# 광화문: 37.57617987638928, 126.97694467807803
# 영풍문고-종로본점: 37.56999947440024, 126.98228391170936
from folium.features import CustomIcon

icon_img = "batman.png"
icon = CustomIcon(icon_img, icon_size=(45,45))

seoul_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=15)
popup1 = folium.Popup("서울특별시청", max_width=200)
popup2 = folium.Popup("덕수궁", max_width=200)
popup3 = folium.Popup("서울지방경찰청", max_width=200)
popup4 = folium.Popup("광화문", max_width=200)
popup5 = folium.Popup("영풍문고-종로본점", max_width=200)
folium.Marker(location=[37.56652479270633, 126.9779129116262], popup=popup1, icon=folium.Icon(color="red", icon="star")).add_to(seoul_folium)
folium.Marker(location=[37.56585320879458, 126.97514572197248], popup=popup2, icon=folium.Icon(color="pink", icon="ok")).add_to(seoul_folium)
folium.Marker(location=[37.57504045678939, 126.97195576938117], popup=popup3, icon=folium.Icon(color="orange", icon="user")).add_to(seoul_folium)
folium.Marker(location=[37.57617987638928, 126.97694467807803], popup=popup4, icon=icon).add_to(seoul_folium)
folium.Marker(location=[37.56999947440024, 126.98228391170936], popup=popup5).add_to(seoul_folium)
seoul_folium

# 지도: 
```

```python
# CircleMarker() 원의 범위를 마커로 생성
seoul_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=15)
popup1 = folium.Popup("서울특별시청", max_width=200)
folium.CircleMarker(location=[37.56652479270633, 126.9779129116262], popup=popup1, radius=100, color="red", fill_color="#EC4074").add_to(seoul_folium)
seoul_folium
```

```python
# CircleMarker() 원의 범위를 마커로 생성
map = folium.Map(location=[37.58,127.0],zoom_start=11)

def select_color(row):
    year = int(row["개관일"].split('-')[0])
    if year > 2010:
        return "blue"
    else:
        return "red"

for _, row in library.iterrows():
    popup = folium.Popup(row["시설명"], max_width=200)
    folium.Circle(location=[row["위도"], row["경도"]], radius=200, color=select_color(row), fill_color=select_color(row)).add_to(map)

map
```
