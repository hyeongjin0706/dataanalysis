```python
# 2. 버거킹은 맥도날드의 매장수를 넘어섰을까? (2023-05현재)

# 맥도날드 데이터 프레임에 저장
df_mc = df_korea_copy.loc[df_korea_copy["상호명"].str.contains(r'맥도널드|맥도날드')]
```

```python
# 버거킹 데이터 프레임에 저장
df_burgerking = df_korea_copy.loc[df_korea_copy["상호명"].str.contains('버거킹')]
```

```python
# '분류' 파생변수 생성 후, 각각 값 넣어줌
df_mc["분류"] = "맥도날드"
df_burgerking["분류"] = "버거킹"
```

```python
# 버거킹과 맥도날드 데이터 합치기
df_mc_df_burgerking = pd.concat([df_mc, df_burgerking], axis=0)
df_mc_df_burgerking
```

```python
# 맥도날드 버거킹 총 갯수를 뽑아서 다시 대입
df_mc_df_burgerking = pd.DataFrame(df_mc_df_burgerking["분류"].value_counts())
df_mc_df_burgerking
```

```python
# 인덱스를 리셋해줌
df_mc_df_burgerking = df_mc_df_burgerking.reset_index()
df_mc_df_burgerking
```

```python
# 컬럼명 재지정
df_mc_df_burgerking.columns= ["분류", "매장 수"]
df_mc_df_burgerking
```

```python
# 막대 그래프로 버거킹 매장 수와 맥도날드 매장 수를 표현
# 맥도날드의 매장수가 많다는 것을 알 수 있음
plt.figure(figsize=(12,8))
plt.xticks(rotation=45)
sns.barplot(data=df_mc_df_burgerking, x="분류", y="매장 수")
```

```python
# 3. 패스트푸드 매장(롯데리아, 맥도날드, 버거킹, 맘스터치)은 어느 지역에 많이 분포할까? (시도)
# df_fastfood에 상호명으로 분류한 패스트푸드 매장  저장
df_fastfood = df_korea_copy.loc[df_korea_copy["상호명"].str.contains(r'롯데리아|맥도날드|맥도널드|버거킹|맘스터치')]
df_fastfood
```

```python
# '시도명'을 기준으로 그룹화하고, 그룹별로 패스트 푸드 매장의 수를 구해서 df_fastfood_by_province에 저장
df_fastfood_by_province = df_fastfood.groupby('시도명').size().reset_index(name='매장 수')
df_fastfood_by_province
```

```python
# 시도별 매장수를 보기쉽게 표현하기 위해 내림차순 후, 재저장
df_fastfood_by_province = df_fastfood_by_province.sort_values(by="매장 수")
```

```python
# 막대그래프로 지역별 패스트푸드 매장 수를 표현
# 경기도에 가장 많이 분포하는 사실을 알 수 있음
plt.figure(figsize=(12,8))
plt.xticks(rotation=45)
sns.barplot(data=df_fastfood_by_province, x="시도명", y="매장 수")
```
